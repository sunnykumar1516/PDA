# creating a new column based on imdb rating
# rating from 1-3 : flop movie
# rating from 4-5 : average movie
# rating from 6-8 : hit
# rating from 9-10 : blockbuster


```python
def getMovie_category(d):
    if d>=0 and d<=3.5:
        return "flop"
    elif d>3.5 and d<= 6.5:
        return "average"
    elif d>6.5 and d<= 8.5:
        return "hit"
    elif d>8.5 and d<=10:
        return "blockbuster"
    
```

```python
df['imdb_score']
```

```python
df["movie_category"]= df["imdb_score"].apply(getMovie_category)
```
